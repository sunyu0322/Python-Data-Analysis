# 项目：评估和清理英国电商公司销售数据

## 分析目标

此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 简介

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：
- `InvoiceNo`: 发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
- `StockCode`: 产品代码。5位数，作为产品的唯一标识符。
- `Description`: 产品名称。
- `Quantity`: 产品在交易中的数量。
- `InvoiceDate`: 发票日期和时间。交易发生的日期和时间。
- `UnitPrice`: 单价。价格单位为英镑（£）。
- `CustomerID`: 客户编号。5位数，作为客户的唯一标识符。
- `Country`: 国家名称。客户所居住的国家的名称。

## 评估缺失数据
在了解了```'Description'```存在缺失值后，根据条件提取出缺失观察值

````python
df1[df1['Description'].isnull()]
````

有1454行的交易数据缺失```''Description''```的值
从输出结果来看，这些缺失```''Description''```的交易数据，```'UnitPrice'```都为0。
为了验证猜想，我们增加筛选条件，看是否存在```''Description''```变量缺失且```'UnitPrice'```不为0的数据

````python
df1[(df1['Description'].isnull())&('UnitPrice' != 0 )]
````

筛选出来的结构数量为0，说明缺失```''Description''```的值，同时也不具备有效的```'UnitPrice'```值。
前者代表商品名称，后者是单价，这两者缺失的数据显然也没有有效意义，认为可以后续被删除。

````python
df1[df1['CustomerID'].isnull()]
````

```'CustomerID'```仅仅代表客户编号，不是分析商品的必要变量。并且从输出结果来看，有些CustomerID缺失的交易数据仍然有效，因此保留此变量为空的观察值。

## 清理数据
要做的事项有：
1. 改数据类型```'InvoiceDate'```应该是日期、```'CustomerID'```应该是字符串
2. 删除缺失```''Description''```的值
3.  ```'USA'``` 统一为 ```'United States'```  都表示美国，
4. ```'UK'``` 和 ```'U.K.'```统一为```'United Kingdom'``` 都是英国
5. ```'Quantity'```为负数的商品都可以删除
6. ```df1['UnitPrice']<0```的观察值擅长

````python
clean_data = df1.copy()
clean_data
````

````python
# 转换时间类型
clean_data['InvoiceDate'] = pd.to_datetime(clean_data['InvoiceDate'])
clean_data['InvoiceDate']
````

```'CustomerID'```转换类型为字符串

````python
clean_data['CustomerID'] = clean_data['CustomerID'].astype(str)
clean_data['CustomerID'] 
````

````python
# 为了删除结尾的.0，用到Python Pandas Series.str.slice()用法
clean_data['CustomerID'] = clean_data['CustomerID'].str.slice(0,-2)
clean_data['CustomerID']
````

删除缺失```''Description''```的值

````python
clean_data.dropna(subset=['Description'],inplace=True)
````

````python
clean_data['Description'].isnull().sum()
````

````python
clean_data['Country']=clean_data['Country'].replace({'USA':'United States'})
````

````python
len(clean_data[clean_data['Country']=='USA'])
````

把```'UK'``` 和 ```'U.K.'```统一为```'United Kingdom'``` 

````python
clean_data['Country']=clean_data['Country'].replace({'UK':'United Kingdom','U.K.':'United Kingdom'})
````

````python
print(len(clean_data[clean_data['Country']=='UK']))
print(len(clean_data[clean_data['Country']=='U.K.']))
````

5. ```'Quantity'```为负数的商品都可以删除
6. ```df1['UnitPrice']<0```的观察值删除

````python
clean_data=clean_data[clean_data['Quantity']>=0]
````

````python
clean_data[clean_data['Quantity']<0]
````

````python
clean_data=clean_data[clean_data['UnitPrice']>=0]
````

````python
clean_data[clean_data['UnitPrice']<0]
````

````python
clean_data
````
